```python
def missing_values_percent(dataframe):
    missing_values = dataframe.isna().sum()
    percentage_missing = (missing_values / len(dataframe) * 100).round(2)
    result_movie = pd.DataFrame({'Missing Values': missing_values, "Percentage": percentage_missing})
    result_movie["Percentage"] = result_movie["Percentage"].astype(str) + '%'
    
    return result_movie

result = missing_values_percent(heartratedata)
result
```

```python
bldpList = []
systolicbp = []
diastolicbp = []

for data in bldp:
    tempdata = data.split("/")
    systolicbp.append(tempdata[0])
    diastolicbp.append(tempdata[1])
```

```python
len(systolicbp)
```

```python
len(diastolicbp)
type(systolicbp[0])
```

```python
bprankList = []

def getRange1(num):
    if num >= 130 and num <= 139:
        return True
    
def getRange2(num):
    if num >= 80 and num <= 89:
        return True

for i in range(0, len(diastolicbp)):
    if int(systolicbp[i]) < 120 and int(diastolicbp[i]) < 80:
        bprankList.append(1)
    elif int(systolicbp[i]) >= 120 and int(systolicbp[i]) <= 129 and int(diastolicbp[i]) < 80:
        bprankList.append(2)
    elif getRange1(int(systolicbp[i])) or getRange2(int(diastolicbp[i])):
        bprankList.append(3)
    elif int(systolicbp[i]) >= 140 or int(diastolicbp[i]) >= 90:
        bprankList.append(4)

len(bprankList)
```

```python
bprankList
```

```python
heartratedata["BPrank"] = bprankList
```

```python
# Age, Cholestrol, Heart Rate
AgeMean = np.mean(list(heartratedata["Age"]))
AgeStd = np.std(list(heartratedata["Age"]))
ChoMean = np.mean(list(heartratedata["Cholesterol"]))
ChoStd = np.std(list(heartratedata["Cholesterol"]))
HRMean = np.mean(list(heartratedata["Heart Rate"]))
HRStd = np.std(list(heartratedata["Heart Rate"]))
```

```python
def getNormal(mean, std, arr):
    tempArr = []
    for data in arr:
        temp = (data - mean) / std
        tempArr.append(temp)
    return tempArr

AgeNormalList = getNormal(AgeMean, AgeStd, list(heartratedata["Age"]))
ChoNormalList = getNormal(ChoMean, ChoStd, list(heartratedata["Cholesterol"]))
HRNormalList = getNormal(HRMean, HRStd, list(heartratedata["Heart Rate"]))

AgeNormalList
```
